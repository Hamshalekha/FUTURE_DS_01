```python
#Load the Dataset

import pandas as pd

df = pd.read_csv("WA_Fn-UseC_-Telco-Customer-Churn.csv")

df.head()
df.info()
```

```python
# Data Cleaning
# Check missing values
df.isnull().sum()
```

```python
# Check blank spaces in TotalCharges
(df['TotalCharges'] == ' ').sum()
```

```python
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
```

```python
df['TotalCharges'].isnull().sum()
```

```python
df = df.dropna()
```

```python
df.info()
```

```python
df.duplicated().sum()
```

```python
# Total Customers
total_customers = len(df)
print("Total Customers:", total_customers)

# Churned Customers
churned = len(df[df['Churn'] == 'Yes'])
print("Churned Customers:", churned)

# Retained Customers
retained = len(df[df['Churn'] == 'No'])
print("Retained Customers:", retained)

# Churn Rate
churn_rate = (churned / total_customers) * 100
print("Churn Rate:", round(churn_rate,2), "%")

# Retention Rate
retention_rate = (retained / total_customers) * 100
print("Retention Rate:", round(retention_rate,2), "%")
```

```python
print("Average Monthly Charges:", round(df['MonthlyCharges'].mean(),2))

print("Average Customer Lifetime (Months):", round(df['tenure'].mean(),2))

print("Total Revenue:", round(df['TotalCharges'].sum(),2))
```
